```python
medications = ["Piroxicam", "Warfarin", "Metformin", "Aspirin"]
#index:              0            1           2            3

print(medications[0])   # Piroxicam  ← first item
print(medications[1])   # Warfarin
print(medications[-1])  # Aspirin    ← last item (negative counts from end)
print(medications[-2])  # Metformin  ← second to last
```

```python
medications = ["Piroxicam", "Warfarin", "Metformin", "Aspirin"]

print(medications[0:2])   # ['Piroxicam', 'Warfarin']   index 0 and 1
print(medications[1:3])   # ['Warfarin', 'Metformin']   index 1 and 2
print(medications[2:4])   # ['Metformin', 'Aspirin']     from index 2 to index 3
print(medications[:2])    # ['Piroxicam', 'Warfarin']    from start to index 1
```
