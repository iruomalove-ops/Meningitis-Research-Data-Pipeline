```python
#return values from a function
def calculate_bmi(weight_kg, height_cm):
    height_m = height_cm / 100  # convert height to meters
    bmi = weight_kg / (height_m ** 2)  # BMI formula
    return bmi

#use the function to calculate BMI for a volunteer
volunteer_1_bmi = calculate_bmi(70, 175)
volunteer_2_bmi = calculate_bmi(90, 165)
volunteer_3_bmi = calculate_bmi(55, 170)

print(f"volunteer 1 BMI: {volunteer_1_bmi:.2f}")
print(f"volunteer 2 BMI: {volunteer_2_bmi:.2f}")
print(f"volunteer 3 BMI: {volunteer_3_bmi:.2f}")
```
